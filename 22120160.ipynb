Tìm phần tử khác không đầu tiên trong hàng:</span> Tìm phần tử khác không đầu tiên trong hàng: Đối với mỗi hàng của ma trận, hàm này bắt đầu bằng cách tìm kiếm phần tử khác không đầu tiên. Cụ thể, nó sẽ chia các phần tử dưới đường chéo chính cho phần tử chính nằm trên đường chéo chính và sau đó trừ tích của phần tử này với mỗi phần tử trong hàng hiện tại. Quá trình này tiếp tục cho đến khi tất cả các phần tử dưới đường chéo chính đều là 0.<div>

    <span style="font-weight: bold">4. Sắp xếp lại ma trận dựa trên số lượng phần tử bằng 0 trong mỗi hàng: </span> Cuối cùng, ma trận được sắp xếp lại dựa trên số lượng phần tử bằng 0 trong mỗi hàng. Điều này giúp tối ưu hóa quá trình tìm kiếm phần tử khác không đầu tiên trong các bước tiếp theo của thuật toán.<div>

```python
import sympy as sy
def count_zeros(row):
    return sum(1 for element in row if element == 0)
def Gauss_elimination(A):
    n = len(A)
    m = len(A[0])  # Số cột của ma trận
    rankA = 0  # Bậc của ma trận, ban đầu đặt là 0


    for i in range(n):  
        # Tìm hàng chứa phần tử khác không đầu tiên
        for j in range(m):
            if A[i][j] != 0:
                # Đổi hàng i và hàng rankA
                A[rankA], A[i] = A[i], A[rankA]
                break
        else:
            # Nếu không tìm thấy phần tử khác không, tiếp tục với hàng tiếp theo
            continue

        # Loại bỏ các phần tử dưới đường chéo chính
        for j in range(rankA + 1, n):
            # Kiểm tra điều kiện trước khi thực hiện phép chia
            if A[rankA][rankA] != 0:
                factor = A[j][rankA] / A[rankA][rankA]
                for k in range(rankA, m):
                    A[j][k] -= factor * A[rankA][k]

        # Tăng bậc lên 1 sau mỗi bước khử
        rankA += 1
    
    A.sort(key=count_zeros)

    return A, rankA
```

```python
import sympy as sy

def infiniteCase(matrix, n, m, rankA):
    FreeVariables = n - rankA
    # Initialize a set of uncalculated variables
    uncalculated_variables = set(range(n))
    solution = [[0] * (FreeVariables + 1) for _ in range(n)]
    freeVarPos = 0

    for row in range(m - 1, -1, -1):
        if all(element == 0 for element in matrix[row]):  
            continue
        else:
            # Find the first non-zero element in the row
            firstNonZero = next(i for i, x in enumerate(matrix[row]) if x)
            constVal = matrix[row][n]
            solution[firstNonZero] = [constVal] + [0 for _ in range(FreeVariables)]
            
            
            for i in range(n - 1, firstNonZero, -1):
                if i in uncalculated_variables:
                    freeVarPos += 1
                    uncalculated_variables.remove(i)
                    # mark the free variable in the solution
                    #print(i, freeVarPos)
                    solution[i][freeVarPos] = 1
                    
                # Initialize an empty list
                new_solution = []
                # Iterate over corresponding elements in the two lists
                for sj, si in zip(solution[firstNonZero], [x*matrix[row][i] for x in solution[i]]):
                    # Subtract the elements and append the result to the new list
                    new_solution.append(sj - si)
                # Assign the new list to solution[firstNonZero]
                solution[firstNonZero] = new_solution
                
            uncalculated_variables.remove(firstNonZero)
            
    free_variables = []
    dependent_variables = []
    
    for i in range(n):
        if i in uncalculated_variables:  # If the variable is free
            if all(coeff == 0 for coeff in solution[i][1:]):  # Check if all coefficients are 0
                free_variables.append({'name': f't_{i}', 'coefficients': [1]})
            else:
                free_variables.append({'name': f't_{i}', 'coefficients': solution[i][1:]})
        else:
            dependent_variables.append({'name': f'x_{i + 1}', 'coefficients': solution[i][1:]})

    return free_variables, dependent_variables
```

```python
import numpy as np
def back_substitution(A, rankAb, rankA):
    numbersOfUnknowns = len(A[0]) - 1
    numbersOfEquations = len(A)
    x = [0] * numbersOfUnknowns
    if rankA < rankAb:
        print("Hệ phương trình vô nghiệm.\n")
    elif rankA == rankAb and rankAb < numbersOfUnknowns:
        print("Hệ phương trình có vô số nghiệm.")
        freeVar, dependentVar = infiniteCase(A, numbersOfUnknowns, numbersOfEquations, rankAb)
        printInfiniteSolution(freeVar, dependentVar)
    else:
        for i in range(numbersOfUnknowns - 1, -1, -1):    
            x[i] = A[i][numbersOfUnknowns] / A[i][i]
            for k in range(i - 1, -1, -1):
                A[k][numbersOfUnknowns] -= A[k][i] * x[i]
        res = [sy.nsimplify(i) for i in x]
        print("Hệ phương trình có tập nghiệm duy nhất:")
        for i in range(len(res)):
            print(f"x_{i+1} = {sy.simplify(res[i])}")
```

```python
import sympy as sy

def input_matrix_and_vector():
    print("Nhập ma trận A:")
    A = []
    print("Nhập vào kích thước của ma trận A.\n")
    n=int(input("Nhập vào số dòng của ma trận A: "))
    m=int(input("Nhập vào số cột của ma trận A: "))
    print("Ma trận có",n,"dòng và",m,"cột.\n" )
    for i in range(1, n+1):
        row = []
        for j in range(1, m+1):
            entry = float(input(f"Nhập phần tử a_{i}{j}: "))
            row.append(entry)
        A.append(row)
    
    print("Nhập vector b:")
    b = []
    for i in range(1, n+1):
        entry = float(input(f"Nhập phần tử b_{i}: "))
        b.append(entry)

    return A, b

def augment_matrix(A, b):
    # Thêm vector b vào ma trận A để tạo ma trận bổ sung [A|b]
    augmented_matrix = []
    for i in range(len(A)):
        augmented_matrix.append(A[i] + [b[i]])
    return augmented_matrix

def printMatrix(A):
    #print("The matrix after Gaussian elimination:")
    for i in range(len(A)):
        row = [sy.nsimplify(x, tolerance=1e-10) for x in A[i]]
        print(row)
        
def print_vector(vector):
    print("[", end="")
    for i in range(len(vector)):
        print(vector[i], end="")
        if i < len(vector) - 1:
            print(", ", end="")
    print("]")
    
def count_zeros(row):
    return sum(1 for element in row if element == 0)
def Gauss_elimination(A):
    n = len(A)
    m = len(A[0])  # Số cột của ma trận
    rankA = 0  # Bậc của ma trận, ban đầu đặt là 0


    for i in range(n):  
        # Tìm hàng chứa phần tử khác không đầu tiên
        for j in range(m):
            if A[i][j] != 0:
                # Đổi hàng i và hàng rankA
                A[rankA], A[i] = A[i], A[rankA]
                break
        else:
            # Nếu không tìm thấy phần tử khác không, tiếp tục với hàng tiếp theo
            continue

        # Loại bỏ các phần tử dưới đường chéo chính
        for j in range(rankA + 1, n):
            # Kiểm tra điều kiện trước khi thực hiện phép chia
            if A[rankA][rankA] != 0:
                factor = A[j][rankA] / A[rankA][rankA]
                for k in range(rankA, m):
                    A[j][k] -= factor * A[rankA][k]

        # Tăng bậc lên 1 sau mỗi bước khử
        rankA += 1
    
    A.sort(key=count_zeros)

    return A, rankA

def printInfiniteSolution(free_vars, dependent_vars):
    res = []

    for var in free_vars:
        coefficients = var['coefficients']
        variables = [f't_{i+1}' for i in range(len(coefficients))]
        terms = [f'{coeff}*{var}' for coeff,
                 var in zip(coefficients, variables) if coeff != 0]
        equation = var['name'] + ' = ' + ' + '.join(terms)
        res.append(equation)

    for var in dependent_vars:
        coefficients = var['coefficients']
        variables = [f't_{i+1}' for i in range(len(coefficients))]
        terms = [f'{coeff}*{var}' for coeff,
                 var in zip(coefficients, variables) if coeff != 0]
        equation = var['name'] + ' = ' + ' + '.join(terms)
        res.append(equation)

    for i, result in enumerate(res):
        print(result)

    print("For t_i is any real number for i = ", end='')
    for i in range(0, len(coefficients)):
        if i != len(coefficients) - 1:
            print(i+1, end=', ')
        else: 
            print(i+1)
            

def infiniteCase(matrix, n, m, rankA):
    FreeVariables = n - rankA
    # Initialize a set of uncalculated variables
    uncalculated_variables = set(range(n))
    solution = [[0] * (FreeVariables + 1) for _ in range(n)]
    freeVarPos = 0

    for row in range(m - 1, -1, -1):
        if all(element == 0 for element in matrix[row]):  
            continue
        else:
            # Find the first non-zero element in the row
            firstNonZero = next(i for i, x in enumerate(matrix[row]) if x)
            constVal = matrix[row][n]
            solution[firstNonZero] = [constVal] + [0 for _ in range(FreeVariables)]
            
            
            for i in range(n - 1, firstNonZero, -1):
                if i in uncalculated_variables:
                    freeVarPos += 1
                    uncalculated_variables.remove(i)
                    # mark the free variable in the solution
                    #print(i, freeVarPos)
                    solution[i][freeVarPos] = 1
                    
                # Initialize an empty list
                new_solution = []
                # Iterate over corresponding elements in the two lists
                for sj, si in zip(solution[firstNonZero], [x*matrix[row][i] for x in solution[i]]):
                    # Subtract the elements and append the result to the new list
                    new_solution.append(sj - si)
                # Assign the new list to solution[firstNonZero]
                solution[firstNonZero] = new_solution
                
            uncalculated_variables.remove(firstNonZero)
            
    free_variables = []
    dependent_variables = []
    
    for i in range(n):
        if i in uncalculated_variables:  # If the variable is free
            if all(coeff == 0 for coeff in solution[i][1:]):  # Check if all coefficients are 0
                free_variables.append({'name': f't_{i}', 'coefficients': [1]})
            else:
                free_variables.append({'name': f't_{i}', 'coefficients': solution[i][1:]})
        else:
            dependent_variables.append({'name': f'x_{i + 1}', 'coefficients': solution[i][1:]})

    return free_variables, dependent_variables

def back_substitution(A, rankAb, rankA):
    numbersOfUnknowns = len(A[0]) - 1
    numbersOfEquations = len(A)
    x = [0] * numbersOfUnknowns
    if rankA < rankAb:
        print("Hệ phương trình vô nghiệm.\n")
    elif rankA == rankAb and rankAb < numbersOfUnknowns:
        print("Hệ phương trình có vô số nghiệm.")
        freeVar, dependentVar = infiniteCase(A, numbersOfUnknowns, numbersOfEquations, rankAb)
        printInfiniteSolution(freeVar, dependentVar)
    else:
        for i in range(numbersOfUnknowns - 1, -1, -1):    
            x[i] = A[i][numbersOfUnknowns] / A[i][i]
            for k in range(i - 1, -1, -1):
                A[k][numbersOfUnknowns] -= A[k][i] * x[i]
        res = [sy.nsimplify(i) for i in x]
        print("Hệ phương trình có tập nghiệm duy nhất:")
        for i in range(len(res)):
            print(f"x_{i+1} = {sy.simplify(res[i])}")
            
if __name__ == "__main__":            
    A, b = input_matrix_and_vector()
    print("Ma trận A: ") 
    printMatrix(A)
    print("Vector b : ") 
    print_vector(b)
    matrix = augment_matrix(A, b)
    print("Ma trận bổ sung [A|b]: ")
    printMatrix(matrix)

    row_echelon_form_A, rankA = Gauss_elimination(A)

    matrix, rankAb = Gauss_elimination(matrix)
    print("Bậc của ma trận: ", rankAb)
    print("Ma trận sau khi dùng phép khử Gauss:")
    for i in range(len(matrix)):
        row = [sy.nsimplify(x, tolerance=1e-10) for x in matrix[i]]
        print(row)

    back_substitution(matrix, rankAb, rankA)
```
